```python
import pandas as pd

def getStat(inst, group):
    df = pd.read_csv("./status/{}.csv".format(group))
    statuses = []
    for i in inst:
        statuses.append(df[i][0])
    return statuses

print("Ya")
    
```

```python
import math
def satpurity(inst, group):
    stat = getStat(inst, group)
    sat = 0
    unsat = 0
    for i in stat:
        if i == "SAT":
            sat += 1
        if i == "UNSAT":
            unsat += 1
    p = sat / (sat + unsat) if sat + unsat > 0 else 0

        
    entropy = -p * math.log2(p) - (1 - p) * math.log2(1 - p) if p not in [0, 1] else 0
    return entropy, sat + unsat
```

```python
import pandas as pd
import numpy as np
from utils import *
import warnings
import pickle

warnings.filterwarnings("ignore")


minsize = 8
def split(a, instances):
        keys = instances

        if (len(instances) < 2 * minsize):
            return SplitNode(hashes=keys, config=getBestSet(a, keys)[0])

        best = 400000
        bestindex = 0
        bsplit = 0
        bestconf = (0,0)
        blower = []
        bupper = []
        for i in range(56):
            n, nsplit, conf, lower, upper = evaluatemeta(i, keys, a, minsize)
            if n < best:
                best = n
                bsplit = nsplit
                bestindex = i
                bestconf = conf
                blower = lower
                bupper = upper

        

        if best / getBestSet(a, keys)[1] < 0.9:
            vars = []
            feat = getFeat(keys)

            for key in keys:
                line = (feat[feat["hash"] == key])[feat.keys()[bestindex + 1]]

                vars.append((key, float(line)))
            vars.sort(key=lambda x: x[1])
            return SplitNode(split_var=bestindex, split_point = (vars[len(blower)][1] - vars[len(blower) - 1][1]) / 2.0, left=split(a, [x[0] for x in blower]), right=split(a, [x[0] for x in bupper]))
        else:
            return SplitNode(hashes=keys, config=getBestSet(a, keys)[0])



splits = []


for a in range(35):

    ordered_df = pd.read_csv('ordered/{}.csv'.format(a))
    # Get the list of all keys (column names)
    keys = ordered_df.keys()[1:]
    splits.append(split(a, keys))
    print(a)

with open('splits_minsize_{}.pkl'.format(minsize), 'wb') as outp:
    pickle.dump(splits, outp, pickle.HIGHEST_PROTOCOL)

```

```python
def flatten(t):
    if t.is_leaf():
        return [(t.config, t.hashes)]
    else:
        return flatten(t.left) + flatten(t.right)

with open('splits_minsize_8.pkl', 'rb') as inp:
    tr = pickle.load(inp)
    all = []
    for i,t in enumerate(tr):
        all += [(i,flatten(t))]

    alldic = {}
    for a in all:
        group = a[0]
        for leaf in a[1]:
            conf =  str(getConf(group, leaf[0]))
            if conf in alldic.keys():
                alldic[conf].extend(leaf[1].copy())
            else:
                alldic[conf] = leaf[1].copy()
    with open('splits_8_dic.txt', 'w') as f:
        for key, value in alldic.items():
            f.write(f"{key}: {', '.join(value)}\n")
```
