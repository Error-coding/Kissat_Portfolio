```python
import pandas as pd

df = pd.read_csv('top40.csv')

configs = df['configuration'].drop_duplicates().tolist()

allargs = []

for c in configs:
    if c == "Default":
        continue
    c = c[1:-1]
    c = c.split(", ")
    args = {}

    for arg in c:
        name, val = arg.split(": ")
        args[name[1:-1]] = val
    allargs.append(args)
        

#header = "Configuration & 6 & 7 & 9 & 16 & 21 & 25 & 27 & 29 \\\\"

for k in allargs[0].keys():
    line = k + "&"
    for i in range(8):
        line += allargs[i][k] + "&"
    line = line[:-1]
    line += "\\\\"
    print(line)
```
